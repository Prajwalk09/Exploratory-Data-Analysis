```python
country_data = pd.read_excel('Country-Code.xlsx')
country_data.head()
```

```python
new_data = data.merge(country_data, on = 'Country Code', how = 'left')
```

```python
new_data.head()
```

```python
new_data.columns
```

```python
country_names = new_data['Country'].value_counts().index
```

```python
# plt.figure(figsize = (12,6))
plt.pie(x = new_data['Country'].value_counts().values[0:3], 
        labels = country_names[0:3],
        autopct = "%1.2f%%")
plt.title("Top 3 Countries")
plt.show()
```

Maximum transactions are from India, followed by USA and The United Kingdom

```python
new_data.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = new_data.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns = {0:'Rating Count'})
```

```python
ratings
```

2000+ People have not given any ratings <br>
Ratings above 4.5 indicates that food was excellent <br>
Ratings between 4.0 and 4.4 indicate that rating is very good <br>
Similarly for other ratings

```python
plt.figure(figsize = (12,6))
sns.barplot(x = ratings['Aggregate rating'], 
            y = ratings['Rating Count'])
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = 'Rating color',
            y = 'Rating Count',
            data = ratings, 
            palette = ['gray','red','orange','yellow','green','green'] )
plt.show()
```

```python
plt.figure(figsize = (12,6))
sns.barplot(x = ratings['Aggregate rating'], 
            y = ratings['Rating Count'],
            hue = ratings['Rating color'],
            palette = ['white','red','orange','yellow','green','green'])
plt.show()
```

```python
sns.countplot(data = ratings,
              x = 'Rating color',
              palette = ['white','red','orange','yellow','green','green'])
```

```python
# Countries with 0 rating
new_data[new_data['Rating color'] == 'White'].groupby('Country').size().reset_index()
```

Most 0 Ratings are from Indian Customers

```python
city_values = new_data['City'].value_counts().values
city = new_data['City'].value_counts().keys()
```

```python
cuisine_values = new_data['Cuisines'].value_counts().head(10).values
cuisines = new_data['Cuisines'].value_counts().head(10).keys()
plt.pie(x = cuisine_values,
        labels = cuisines,
        autopct = '%1.2f')
plt.title("Top 10 Cuisines")
plt.show()
```
